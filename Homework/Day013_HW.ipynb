### ① 印出最新文章的「作者」「標題」「時間」

```python
def getpttdata_new(url):
    r = requests.get(url)

    soup = BeautifulSoup(r.text, "html5lib")

    for d in soup.find_all(class_="title"):
        print(d.text.replace('\t', '').replace('\n', ''))
        try:
            r = BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href']).text, "html5lib")
            metalines = r.find_all(class_='article-metaline')
            print('作者: ', metalines[0].find(class_='article-meta-value').text)
            print('日期: ', metalines[2].find(class_='article-meta-value').text)
            break
        except:
            continue
getpttdata_new('https://www.ptt.cc/bbs/NBA/index.html')
```

### ② 印出第一頁所有文章的「作者」「標題」「時間」

```python
def getpttdata(url):
    r = requests.get(url)

    soup = BeautifulSoup(r.text, "html5lib")

    for d in soup.find_all(class_="title"):
        print(d.text.replace('\t', '').replace('\n', ''))
        try:
            r = BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href']).text, "html5lib")
            metalines = r.find_all(class_='article-metaline')
            print('作者: ', metalines[0].find(class_='article-meta-value').text)
            print('日期: ', metalines[2].find(class_='article-meta-value').text)
        except:
            continue
getpttdata('https://www.ptt.cc/bbs/NBA/index.html')
```

```python
import pandas as pd
columns = ["Author","Category","Title", "Date" ]

def getptttable(url):
    r = requests.get(url,cookies={'over18': '1'})

    soup = BeautifulSoup(r.text, "html5lib")
    articles=[]
    
    for d in soup.find_all(class_="title"):
        #print(d.text.replace('\t', '').replace('\n', ''))
        try:
            r = BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href']).text,"html5lib")
            article = []
            for info in r.find_all(class_='article-meta-value'):
                article.append(info.text)
            articles.append(article)
        except:
            continue
    df = pd.DataFrame(articles, columns = columns)
    
    return df
getptttable('https://www.ptt.cc/bbs/NBA/index.html')
```
